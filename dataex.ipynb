```python
# 지역별 폐업률 원평차트 도출
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('fulldata_.csv', encoding='cp949')

# 시/도 추출 함수 (예: '서울특별시 강남구 역삼동 ...' → '서울특별시')
def extract_sido(address):
    try:
        return address.split()[0]
    except:
        return None

data['시도'] = data['소재지전체주소'].apply(extract_sido)

# 서울, 인천, 경기만 필터링
target_sido = ['서울특별시', '인천광역시', '경기도']
data_sido = data[data['시도'].isin(target_sido)]

# 시도별 전체 사업장 수
total = data_sido.groupby('시도').size().reset_index(name='전체사업장수')

# 시도별 폐업 사업장 수
closed = data_sido[data_sido['영업상태명'] == '폐업'].groupby('시도').size().reset_index(name='폐업사업장수')

# 합치기
result = pd.merge(total, closed, on='시도', how='left')
result['폐업사업장수'] = result['폐업사업장수'].fillna(0)
result['폐업률'] = (result['폐업사업장수'] / result['전체사업장수']) * 100

print(result)

# 파이 차트 시각화
plt.rc('font', family='Malgun Gothic')
plt.figure(figsize=(6,6))
plt.pie(result['폐업률'], labels=result['시도'], autopct='%1.1f%%', startangle=140)
plt.title('서울/인천/경기 폐업률 원형 차트')
plt.axis('equal')
plt.show()
```
